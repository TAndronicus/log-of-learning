```python
# analytic solution
# s - savings rate, y - yield, c - consumption rate, x - earnings x[i] - savings after i years
# let multiplier m = 1 + y
# 1' y > 0
# x[1] = sx, x[n + 1] = mx[n] + sx = sx \sum{i = 0}{n - 1}m^i = sx \frac{1 - m^{n - 1}}{1 - m} = \frac{sx}{y} ((1 + y)^{n - 1} - 1)
# cx[n] = (1 - s)x -> solve for n
# n = 1 + log{y + 1}{1 + y\frac{1 - s}{cs}}
# 2' y = 0
# x[1] = sx, x[n + 1] = x[n] + sx = (n + 1)sx
# cx[n] = (1 - s)x = cnsx
# n = \frac{1 - s}{cs}
import math
def years_to_fi(s, y, c):
    if y == 0:
        return (1 - s) / (s * c)
    else:
        return math.log(1 + y * (1 - s) / (c * s), y + 1)
```

```python
from matplotlib import pyplot as plt
```

```python
y, c = 0, .04
ss = [i * .05 for i in range(2, 15)]
years = [years_to_fi(s, y, c) for s in ss]
plt.figure(figsize=(15, 10))
plt.xlabel('Savings rate')
plt.ylabel('Years to FI')
plt.plot(ss, years)
```
